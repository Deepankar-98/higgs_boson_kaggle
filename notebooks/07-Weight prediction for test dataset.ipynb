```python
# For colab:
# df = pd.read_csv('test_MICE_median_ind_QuantileTransformer.csv')

# For local
df1 = pd.read_csv('../data/interim/test_MICE_median_ind_QuantileTransformer.csv')
df1.head()
```

## Dropping un-important columns:

```python
# Dropping the less important feature columns
drop_cols = ['PRI_jet_leading_phi',
 'PRI_jet_subleading_phi',
 'PRI_tau_phi',
 'PRI_met_phi',
 'PRI_lep_phi']

df = df1.drop(columns=drop_cols)
```

## Loading saved model:

```python
import pickle

with open('../models/regg/grid_search_xgb_with_outliers.pkl', 'rb') as fp:
    xgb = pickle.load(fp)
```

## Predict Weight:

```python
weight = xgb.predict(df)
df1['Weight'] = weight
```
